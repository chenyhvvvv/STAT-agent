```python
import scanpy as sc
```

```python
adata_gene = sc.read_h5ad("/import/home3/yhchenmath/Code/spatialdata-agent-data-prepare/demo_data/multi_omics/coad/colon_adenocarcinoma.h5ad")
adata_protein = sc.read_h5ad("/import/home3/yhchenmath/Code/spatialdata-agent-data-prepare/demo_data/multi_omics/coad/colon_adenocarcinoma_protein.h5ad")
```

```python
# Fill NaN with "Unknown"
adata_gene.obs["annotation"] = adata_gene.obs["annotation"].astype("str").replace("nan", "Unknown")
adata_protein.obs["annotation"] = adata_protein.obs["annotation"].astype("str").replace("nan", "Unknown")
```

```python
adata_gene.obs["celltype"] = adata_gene.obs["annotation"]
adata_protein.obs["celltype"] = adata_protein.obs["annotation"]
```
